```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
y1 = np.linspace(10, 100, 15)
y2 = y1 ** 2
y3 = y1 ** (3/2)
```

```python
y1
```

```python
x = np.linspace(1, 15, 15)
x
```

```python
plt.figure(figsize=(12, 7))

plt.title('My Plot')
plt.xlabel('Speed')
plt.ylabel('Distance')
plt.plot(x, y1, color='r', label='y1 (red)', marker='X')
plt.plot(x, y2, color='b', label='y2 (blue)', marker='*')
plt.plot(x, y3, color='g', label='y3 (green)', marker='+')

plt.legend()
plt.savefig('Linegraph.png')
plt.show()
```

```python
cat = ['buff', 'veg', 'chicken']
data = [80, 40, 60]
explosion = (0, 0.1, 0.05)

custom_colours = ['red', 'hotpink', '#4DAF42']
plt.pie(x=data, labels=cat, colors=custom_colours, autopct='%.1f%%', explode=explosion)
plt.savefig('piechart.png')
plt.show()
```

```python
plt.bar(cat, data, color=custom_colours)
plt.savefig('Bar graph.png')
plt.show()
```
