```python
from sklearn.preprocessing import StandardScaler
```

```python
to_scale = ['temp', 'humidity', 'windspeed']

for each_feature in to_scale:
    scaler = StandardScaler()
    df[each_feature] = scaler.fit_transform(df[each_feature].values.reshape(-1, 1))
```
