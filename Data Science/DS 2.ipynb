```python
import matplotlib.pyplot as plt
```

```python
np.random.seed(666)
```

```python
uniform = np.random.uniform(1, 100, 1000) # min, max, size
uniform
```

```python
plt.hist(x=uniform, bins=20)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
plt.show()
```

```python
meu = 1 # mean
sigma = 3 # standard deviation
normal = np.random.normal(meu, sigma, size=1000)
normal
```

```python
plt.hist(x=normal, bins=20)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
plt.show()
```
