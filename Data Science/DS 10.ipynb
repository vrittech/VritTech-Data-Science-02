```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA 
```

```python
from sklearn.decomposition import PCA
```

```python
iris_data = load_iris()
```

```python
X = iris_data.data
y = iris_data.target
```

```python
X[:5]
```

```python
lda = LDA(n_components=2)
pca = PCA(n_components=2)
```

```python
lda_reduced = lda.fit_transform(X, y)
pca_reduced = pca.fit_transform(X)
```

```python
lda_reduced[:5]
```

```python
iris_data.target
```

```python
class_0 = lda_reduced[:50]
class_1 = lda_reduced[50:100]
class_2 = lda_reduced[100:]
```

```python
# class_0
```

```python
plt.scatter(class_0[:, 0], class_0[:, 1], c='r')
plt.scatter(class_1[:, 0], class_1[:, 1], c='b') # np_array[row_operation, column_operation]
plt.scatter(class_2[:, 0], class_2[:, 1], c='g') 
plt.show()
```

```python
iris_data.target_names
```

```python
plt.scatter(lda_reduced[y==0, 0], lda_reduced[y==0, 1], c='r')
plt.scatter(lda_reduced[y==1, 0], lda_reduced[y==1, 1], c='g')
plt.scatter(lda_reduced[y==2, 0], lda_reduced[y==2, 1], c='b')
plt.legend(iris_data.target_names)
plt.show()
```

```python
plt.scatter(pca_reduced[y==0, 0], pca_reduced[y==0, 1], c='r')
plt.scatter(pca_reduced[y==1, 0], pca_reduced[y==1, 1], c='g')
plt.scatter(pca_reduced[y==2, 0], pca_reduced[y==2, 1], c='b')
plt.show()
```
