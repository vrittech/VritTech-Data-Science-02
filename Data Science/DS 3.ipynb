```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
meu = 1 # mean
sigma = 3 # standard deviation
normal = np.random.normal(meu, sigma, size=1000)
normal
```

```python
plt.hist(x=normal, bins=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
plt.show()
```

```python
# minimum val, max val, mean, standard deviation
normal.min()
```

```python
normal.max()
```

```python
normal.mean()
```

```python
normal.std()
```

```python
class StandardScaler:
    def __init__(self, ):
        self.mean = None
        self.std = None
    def fit(self, data):
        self.mean = data.mean()
        self.std = data.std()
    def transform(self, data):
        return (data - self.mean) / self.std
    def fit_transform(self, data):
        self.fit(data)
        return self.transform(data)
```

```python
scaler = StandardScaler()
```

```python
scaled_norm = scaler.fit_transform(normal)
```

```python
normal[:10]
```

```python
scaled_norm[:10]
```

```python
plt.hist(x=normal, bins=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
plt.show()
```

```python
plt.hist(x=scaled_norm, bins=100)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Title')
plt.show()
```

```python
meu = np.array([3, 2])

sigma = np.array([
    [10, 0], # 10 is std for x-axis
    [0, 1]# 1 is std for y-axis
])
```

```python
multi_norm = np.random.multivariate_normal(meu, sigma, 1000)
```

```python
multi_norm[:10]
```

```python
x = multi_norm[:, 0]
y = multi_norm[:, 1]
# multi_norm[row_slicing, column_slicing]
```

```python
plt.scatter(x, y)
plt.show()
```
